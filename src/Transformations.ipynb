```python
# Adjust file paths before running code. This code reads in the files from the disgust folder,
# applies transformations to them which are saved in transformed_images folder.
# Second cell is for downloading transformed images folder from Google Collab
import cv2
import numpy as np
import random
import os


def add_noise(image):
  # Generate Gaussian noise
  mean = 0
  stddev = 0.5 # Standard deviation of the noise
  gaussian_noise = np.random.normal(mean, stddev, image.shape).astype(np.uint8)
  # Add the noise to the image
  noisy_image = cv2.add(image, gaussian_noise)
  noisy_image = cv2.cvtColor(noisy_image, cv2.COLOR_BGR2GRAY) #Change noise to grayscale
  return noisy_image

def flipped_image(image):
  flipped_image = cv2.flip(image, 1) # 1 means horizontal flip
  return flipped_image

def slight_rotation(image):
  #Slight rotation:
  # Get the image dimensions
  height = 48
  width = 48

  # Compute the center of the image
  center = (width // 2, height // 2)

  # Generate a rotation matrix
  M = cv2.getRotationMatrix2D(center, random.randint(5, 40), 1.0) # Rotate by a random amount between 5 to 40 degrees

  rotated_image = cv2.warpAffine(image, M, (width, height))
  return rotated_image

def contrast_and_brightness(image):
  #Add brightness half the time and contrast the other half:
  adjustment = random.choice([True, False])
  if adjustment:
    # Increase contrast
    adjusted_image = cv2.convertScaleAbs(image, alpha=1.25, beta=0)
  else:
    # Increase brightness
    adjusted_image = cv2.convertScaleAbs(image, alpha=1, beta=2)
  return adjusted_image


i = 0 #Counts number of images in disgust folder

#Count the number of images we already have in the disgust folder.
directory = '/content/data/'
for filename in os.listdir(directory):
  file_path = directory + 'im' + str(i) + '.png'
  if os.path.isfile(file_path):  # Check if it's a file
          i = i + 1

print("Disgust folder has " + str(i) + " images")

j = 0

directory = '/content/data/'
for filename in os.listdir(directory):
  file_path = directory + 'im' + str(j) + '.png'
  if os.path.isfile(file_path):  # Check if it's a file
          print(f"Processing file: {file_path}")

          # Load the image
          #transformed_image = cv2.imread('./data/im1.png')
          transformed_image = cv2.imread(file_path)
          transformed_image = add_noise(transformed_image)
          transformed_image = flipped_image(transformed_image)
          transformed_image = slight_rotation(transformed_image)
          transformed_image = contrast_and_brightness(transformed_image)

          # Save or display the noisy image
          cv2.imwrite('/content/transformed_images/im'+ str(j+i) + '.jpg', transformed_image)
          print("image was saved")
          j = j + 1



"""
transformed_image = cv2.imread('./data/im4.png')
#transformed_image = cv2.imread(file_path)
transformed_image = add_noise(transformed_image)
transformed_image = flipped_image(transformed_image)
transformed_image = slight_rotation(transformed_image)
transformed_image = contrast_and_brightness(transformed_image)
cv2.imwrite('/content/im00' + '.jpg', transformed_image)

"""

#Change brightness and contrast
#Padding to "zoom out" on images to create variation in size
#Zooming out then in will degrade the quality so skip for now.
#Skip zoom since images are already zoomed in.

#cv2.imshow('Noisy Image', noisy_image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

```
